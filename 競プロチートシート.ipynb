### アルファベット文字列をN字分ずらす

````python
#元の文字コードにN加えた整数を26で割った余りをord('a')に足すことにより、変換後のコードポイントを算出
def rot_n(s, n):
  #s:文字列 n:何文字ずらすか
    answer = ''
    for letter in s: #文字列を1文字ずつ処理
        answer += chr(ord('a') + (ord(letter)-ord('a')+n) % 26)
    return answer

rot_n('abc', 3)
````

````python
import heapq
class Heapq:
    #最大値を取り出したいときは、desc=Trueにする
    def __init__(self, arr, desc=True):
        if desc:
            arr = [-a for a in arr]
        self.sign = -1 if desc else 1
        self.hq = arr
        heapq.heapify(self.hq)
    #最大or最小を取り出す
    def pop(self):
        return heapq.heappop(self.hq) * self.sign
    #値を追加する
    def push(self, a):
        heapq.heappush(self.hq, a * self.sign)
    #最大or最小を参照するだけ(なくならない)
    def top(self):
        return self.hq[0] * self.sign
````

### 木の直径

＜単純パスの長さの最大値（木の直径）の求め方＞
* 始点全探索した上でDFSとかはTLE
* 最短距離計算を２回やることで求めることが出来る

1. 頂点1から各頂点までの最短距離を求める
2. 最も最短距離が大きかった頂点をuとして、頂点uからの最短距離を求める、そのとき出現した最短距離最大値が木の直径となる

````python
from collections import deque
N = int(input())
G = [[] for _ in range(N)]
for _ in range(N-1):
  a,b = map(int,input().split())
  G[a-1].append(b-1)
  G[b-1].append(a-1)

que = deque()
que.append(0)
dist = [-1]*N
dist[0]=0

while(que):
  v = que.popleft()
  for c in G[v]:
    if dist[c]==-1:
      dist[c] = dist[v]+1
      que.append(c)

tmp = max(dist)
i = dist.index(tmp)

que = deque()
que.append(i)
dist = [-1]*N
dist[i]=0

while(que):
  v = que.popleft()
  for c in G[v]:
    if dist[c]==-1:
      dist[c] = dist[v]+1
      que.append(c)

ans = max(dist)+1
print(ans)
````

## Union-Find

* グループ分けを管理するデータ構造
* グループのリーダーをきめてしまい、リーダーが一致するかどうかで同一グループに所属しているかを確認する
* グループの判定及び結合が可能。分割は不可能
* n個の要素を0～n-1の番号で管理しているので、結合する際は注意する

<用途>
* 閉路判定

````python
from collections import defaultdict

class UnionFind():
    def __init__(self, n):
        self.n = n
        self.parents = [-1] * n
    #要素xが属するグループの根を返す
    def find(self, x):
        if self.parents[x] < 0:
            return x
        else:
            self.parents[x] = self.find(self.parents[x])
            return self.parents[x]
    #要素xが属するグループと要素yが属するグループを結合する
    def union(self, x, y):
        x = self.find(x)
        y = self.find(y)

        if x == y:
            return

        if self.parents[x] > self.parents[y]:
            x, y = y, x

        self.parents[x] += self.parents[y]
        self.parents[y] = x
    #要素xが属するグループの要素数を返す
    def size(self, x):
        return -self.parents[self.find(x)]
    #要素x,yが同じグループに属するかどうかを返す
    def same(self, x, y):
        return self.find(x) == self.find(y)
    #要素xが属するグループに属する要素をリストで返す
    def members(self, x):
        root = self.find(x)
        return [i for i in range(self.n) if self.find(i) == root]
    #すべての根の要素をリストで返す
    def roots(self):
        return [i for i, x in enumerate(self.parents) if x < 0]
    #グループの数を返す
    def group_count(self):
        return len(self.roots())
    #{ルート要素: [そのグループに含まれる要素のリスト], ...}のdefaultdictを返す
    def all_group_members(self):
        group_members = defaultdict(list)
        for member in range(self.n):
            group_members[self.find(member)].append(member)
        return group_members
````

### 閉路判定
方法１：深さ優先探索で解く（割愛）

方法２：UnionFindで解く

**結合した時点でそれぞれの根が一致するかどうかでそのグラフが閉路かどうか決まる**

1. 入力される辺を(Ai,Bi)とする
2. UnionFind木を利用してAi,Biの親（グループ番号）を調べる.それぞれGa,Gbとする
3. Ga=Gbのとき、この辺を貼った時点で閉路が発生すると分かる

````python
#ARC037-B
N,M = map(int,input().split())
uf = UnionFind(N)
heiro = [False]*N
for _ in range(M):
  u,v = map(int,input().split())
  u -= 1
  v -= 1
  #根(親)が一致するとき、閉路になる
  if uf.find(u)==uf.find(v):
    heiro[u] = True
  uf.union(u,v)

a = uf.all_group_members()
ans = len(a)
for k,i in a.items():
    flag = False
    for j in i:
        if heiro[j]==True:
            flag = True
    if flag:
        ans -= 1
print(ans)
````

### 橋

橋：取り除いたときにグラフ全体が非連結になるような辺のこと

橋の本数→全ての辺について、それぞれ橋となるかどうか調べる

＜アルゴリズム＞

1. 全ての辺 e1=(u1,v1) に対して以下を実行する

 * e1 以外の全ての辺 e2=(u2,v2) について以下を実行する

 * u2 と v2 を Union-Find Tree 上の同じ集合にまとめる (unite)

2. u1 と v1 が同じ集合になければ、e1 を除いたグラフは非連結で、e1 は橋

````python
N,M = map(int,input().split())
H = []
#辺を記録
for _ in range(M):
  a,b = map(int,input().split())
  H.append((a-1,b-1))
ans = 0
for i in range(M):#取り除く辺番号
  uf = UnionFind(N)
  for j in range(M):
    if i!=j:
      uf.union(H[j][0],H[j][1])
  #注目する辺以外の全ての同じグループの場合は、その辺は橋ではない
  #注目する辺以外につながる頂点が複数グループに分かれる場合は、その辺は橋である    
  if uf.group_count()!=1:
    ans += 1
print(ans)
````

## セグメント木

セグメントツリーとは

任意の区間について、和・差・or・xor・最小値・最大値を高速に計算できるデータ構造

区間についての計算を木構造を使うことで高速に行うことが出来る

* セグメントツリーの構築:O(N)
* 区間に対する計算:O(logN)
* 値の更新:O(logN)

となる

* 更新後の区間を用いて計算するときは、元の配列ではなく、seg.select(index)を用いて計算すること！
* pythonでは間に合わないのでpypyで提出する！

````python
def segfunc(x,y):
  '''
  区間の
  和→x+y
  差→x-y
  最小値→min(x,y)
  xor→x^y
  or→x|y
  最大公約数→math.gcd(x,y)
  '''
  return x^y

class SegTree:
    '''
    セグメントツリーの構築は
    seg = SegTree(元のリスト、単位元、segfunc)
    単位元：どの数に対して演算を行っても元から変わることがない数
    和：0
    積：1
    or：0
    xor：0
    最小値：大きな数 inf
    最大値：小さな数 -inf

    select(インデックス番号)：値の参照 O(1)
    update(インデックス番号、値)：値の更新O(logN)
    query(左端インデックス番号、右端インデックス番号)：指定区間についての演算結果を返すO(logN)
    '''
    def __init__(self,x_list,init,segfunc):
        self.init=init
        self.segfunc=segfunc
        self.Height=len(x_list).bit_length()+1
        self.Tree=[init]*(2**self.Height)
        self.num=2**(self.Height-1)
        for i in range(len(x_list)):
            self.Tree[2**(self.Height-1)+i]=x_list[i]
        for i in range(2**(self.Height-1)-1,0,-1):
            self.Tree[i]=segfunc(self.Tree[2*i],self.Tree[2*i+1])

    def select(self,k):
        return self.Tree[k+self.num]

    def update(self,k,x):
        i=k+self.num
        self.Tree[i]=x
        while i>1:
            if i%2==0:
                self.Tree[i//2]=self.segfunc(self.Tree[i],self.Tree[i+1])
            else:
                self.Tree[i//2]=self.segfunc(self.Tree[i-1],self.Tree[i])
            i//=2

    def query(self,l,r):
        result=self.init
        l+=self.num
        r+=self.num+1

        while l<r:
            if l%2==1:
                result=self.segfunc(result,self.Tree[l])
                l+=1
            if r%2==1:
                result=self.segfunc(result,self.Tree[r-1])
            l//=2
            r//=2
        return result
````

## 遅延セグメント木

### 区間加算用:[a,b)の値にxを加算する

````python
def segfunc(x,y):
    return x+y
class LazySegTree_RAQ:
    def __init__(self,init_val,segfunc,ide_ele):
        n = len(init_val)
        self.segfunc = segfunc
        self.ide_ele = ide_ele
        self.num = 1<<(n-1).bit_length()
        self.tree = [ide_ele]*2*self.num
        self.lazy = [0]*2*self.num
        for i in range(n):
            self.tree[self.num+i] = init_val[i]
        for i in range(self.num-1,0,-1):
            self.tree[i] = self.segfunc(self.tree[2*i], self.tree[2*i+1])
    def gindex(self,l,r):
        l += self.num
        r += self.num
        lm = l>>(l&-l).bit_length()
        rm = r>>(r&-r).bit_length()
        while r>l:
            if l<=lm:
                yield l
            if r<=rm:
                yield r
            r >>= 1
            l >>= 1
        while l:
            yield l
            l >>= 1
    def propagates(self,*ids):
        for i in reversed(ids):
            v = self.lazy[i]
            if v==0:
                continue
            self.lazy[i] = 0
            self.lazy[2*i] += v
            self.lazy[2*i+1] += v
            self.tree[2*i] += v
            self.tree[2*i+1] += v
    def add(self,l,r,x):
        ids = self.gindex(l,r)
        l += self.num
        r += self.num
        while l<r:
            if l&1:
                self.lazy[l] += x
                self.tree[l] += x
                l += 1
            if r&1:
                self.lazy[r-1] += x
                self.tree[r-1] += x
            r >>= 1
            l >>= 1
        for i in ids:
            self.tree[i] = self.segfunc(self.tree[2*i], self.tree[2*i+1]) + self.lazy[i]
    def query(self,l,r):
        self.propagates(*self.gindex(l,r))
        res = self.ide_ele
        l += self.num
        r += self.num
        while l<r:
            if l&1:
                res = self.segfunc(res,self.tree[l])
                l += 1
            if r&1:
                res = self.segfunc(res,self.tree[r-1])
            l >>= 1
            r >>= 1
        return res
````

### 遅延セグ木（区間更新用）:[a,b)の値をxに更新する

````python
def segfunc(x,y):
    return min(x,y)
class LazySegTree_RUQ:
    def __init__(self,init_val,segfunc,ide_ele):
        n = len(init_val)
        self.segfunc = segfunc
        self.ide_ele = ide_ele
        self.num = 1<<(n-1).bit_length()
        self.tree = [ide_ele]*2*self.num
        self.lazy = [None]*2*self.num
        for i in range(n):
            self.tree[self.num+i] = init_val[i]
        for i in range(self.num-1,0,-1):
            self.tree[i] = self.segfunc(self.tree[2*i],self.tree[2*i+1])
    def gindex(self,l,r):
        l += self.num
        r += self.num
        lm = l>>(l&-l).bit_length()
        rm = r>>(r&-r).bit_length()
        while r>l:
            if l<=lm:
                yield l
            if r<=rm:
                yield r
            r >>= 1
            l >>= 1
        while l:
            yield l
            l >>= 1
    def propagates(self,*ids):
        for i in reversed(ids):
            v = self.lazy[i]
            if v is None:
                continue
            self.lazy[i] = None
            self.lazy[2*i] = v
            self.lazy[2*i+1] = v
            self.tree[2*i] = v
            self.tree[2*i+1] = v
    def update(self,l,r,x):
        ids = self.gindex(l,r)
        self.propagates(*self.gindex(l,r))
        l += self.num
        r += self.num
        while l<r:
            if l&1:
                self.lazy[l] = x
                self.tree[l] = x
                l += 1
            if r&1:
                self.lazy[r-1] = x
                self.tree[r-1] = x
            r >>= 1
            l >>= 1
        for i in ids:
            self.tree[i] = self.segfunc(self.tree[2*i], self.tree[2*i+1])
    def query(self,l,r):
        ids = self.gindex(l,r)
        self.propagates(*self.gindex(l,r))
        res = self.ide_ele
        l += self.num
        r += self.num
        while l<r:
            if l&1:
                res = self.segfunc(res,self.tree[l])
                l += 1
            if r&1:
                res = self.segfunc(res,self.tree[r-1])
            l >>= 1
            r >>= 1
        return res
````

## 順序つき集合

* 集合をソートされた順に保ったまま要素をO(logN)で追加削除行える
* 一般的には、平衡２分探索木というデータ構造で実装される

* x以上の最小の要素の検索
* x以下の値の個数
* 小さい方からk番目の値

などの操作が要素数NとしてO(logN)の計算量でおこなるデータ構造




````python
import math
from bisect import bisect_left, bisect_right, insort
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedMultiset(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedMultiset from iterable. / O(N) if sorted / O(N log N)"
        a = list(a)
        if not all(a[i] <= a[i + 1] for i in range(len(a) - 1)):
            a = sorted(a)
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedMultiset" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def count(self, x: T) -> int:
        "Count the number of x."
        return self.index_right(x) - self.index(x)

    def add(self, x: T) -> None:
        "Add an element. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return
        a = self._find_bucket(x)
        insort(a, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True

    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError

    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return ans
````

### 配列の要素検索

````python
def binary_search(sorted_list, search_value) -> bool:
    """
    二分探索を実施し、検索対象の値がリスト内に含まれるかどうかの
    真偽値を取得する。
    
    Parameters
    ----------
    sorted_list : list of int
        ソート済みのリスト。
    search_value : int
        検索対象の値。

    Returns
    -------
    value_exists : bool
        値がリスト内に含まれるかどうか。
    """
    left_index = 0
    right_index = len(sorted_list) - 1
    while left_index <= right_index:
        middle_index= (left_index + right_index) // 2
        middle_value = sorted_list[middle_index]

        if search_value < middle_value:
            right_index = middle_index - 1
            continue
        if search_value > middle_value:
            left_index = middle_index + 1
            continue

        return True

    return False
````

### DP復元

* DPで答えの復元するには、DPで**逆順に決めていく**ことで実現できる
* ナップサック問題などDPテーブルをTrue/Falseで表さないものでも、後ろから「最適化にできるもの」を選び続ければ良い
* 最短経路問題の復元にも使える
* 復元した後、順番に戻すためにreverseするのを忘れないように!

````python
N,S = map(int,input().split())
A = []
B = []
for _ in range(N):
  a,b = map(int,input().split())
  A.append(a)
  B.append(b)
#配列準備
dp = [[False]*(S+1) for _ in range(N+1)]
dp[0][0] = True #ここミスってた何してんねん
#普通のDP  
for n in range(N):
  for s in range(S+1):
    if dp[n][s]==True:
      if s+A[n]<=S:
        dp[n+1][s+A[n]] = True
      if s+B[n]<=S:
        dp[n+1][s+B[n]] = True

if dp[N][S]==False:
  print('Impossible')
else:
  #ここからDP復元
  #最終日の状態からさかのぼってどの商品を選べばよいかを調べる
  x = N #最終日
  y = S #ピッタシの金額
  ans = ''
  while x>0:
    if y>=A[x-1] and dp[x-1][y-A[x-1]]:#日を一日ずつ下っていく
      ans+='A'
      y -= A[x-1]
    elif y>=B[x-1] and dp[x-1][y-B[x-1]]:
      ans+='B'
      y -= B[x-1]
    x -= 1
  #後ろから復元しているので、reverse()する必要がある
  ans = list(ans)
  ans.reverse()
  ''.join(ans)
  for n in ans:
      print(n,end='')
````

````python
from bisect import bisect_left
def lis(A: list):
    L = [A[0]]
    for a in A[1:]:
        if a > L[-1]:
            # Lの末尾よりaが大きければ増加部分列を延長できる
            L.append(a)
        else:
            # そうでなければ、「aより小さい最大要素の次」をaにする
            # 該当位置は、二分探索で特定できる
            L[bisect_left(L, a)] = a
    return len(L)
````

### いもす法

要素の重複について、「ある連続する区間に、ある数vを足す」という操作をK回繰り返した結果を計算量O(N+K)で高速に計算する方法

* **階差の逆が累積和**であるという性質を利用する

**実装**

1.   **区間[a,b]にvを加算したいとき、a番目の値にvを加算してb+1番目の値に-vを加算する**
2.   加算処理した結果を元に累積和を計算して結果を得る

*   a番目,b番目の部分は問題設定によって変わるので用注意

````python
#一次元いもす法
#数学×アルゴリズム 41問
T = int(input())
N = int(input())
A = [0]*(T+1)
for _ in range(N):
  l,r = map(int,input().split())
  A[l] += 1
  A[r] -= 1 #今回の問題の場合はr+1ではない(WA)

#累積和の配列（効率化済み）
sum = [0]
for i in range(0,len(A)):
  sum.append(sum[i]+A[i])
print(sum)
````

#### 領域加算

* 領域加算を二次元いもす法により行う
* 計算量O(N+HW)

<アルゴリズム>
1. 長方形の左上・右下を-1,左下・右上を+1する(累積和を取るときに、左から右、下から上の方が実装しやすく、その場合はこのように加算する)

* 注意:どの点が重なっているかを求める場合は、左下はそのままの座標を加算、右上はy+=1,x+=1,右下はx+=1 y+=0,左上はy+=1,x+=0とする必要がある
* 逆にどの領域が重なっているかを知りたい場合は、求めたいのは面積なので、+1等の処理はする必要がない
2. 横方向に累積和をとる（左から右）
3. 縦方向に累積和をとる（下から上）

````python
#N枚の長方形の紙があり、左下角と右上角の座標が与えられているとき、
#紙がちょうどk枚重なっている部分の面積についてそれぞれ求める
N = int(input())
#幅は少し多めに見ておくとlist out of rangeを気にしないで済む
grid = [[0]*1005 for _ in range(1005)]
#重みの加算
for _ in range(N):
  #座標のミスに注意する
  lx,ly,rx,ry = map(int,input().split())
  #求めたいのは領域加算の面積
  grid[ly][lx] += 1
  grid[ry][rx] += 1
  grid[ry][lx] -= 1
  grid[ly][rx] -= 1

#縦・横に累積和を取る！

#縦向き
for x in range(1004):
  for y in range(1004):
    grid[y+1][x] += grid[y][x]

#横向き
for y in range(1004):
  for x in range(1004):
    grid[y][x+1] += grid[y][x]

#重なっている枚数別にマスの数を数える
ans = [0]*(N+1)
for y in range(1003):#ここを両方とも1,1003にしたせいで2WA!!!!!!(アホ)
  for x in range(1003):
    #####この書き方！！
    ans[grid[y][x]] += 1

for k in range(1,N+1):
  print(ans[k])
````

### 尺取り法

*   区間の左端と右端を尺取り虫のように動かすことで、条件を満たす区間を高速に見つけるというアルゴリズム
*   O(N^2)がO(N)になる
* dequeを使うと楽に実装できる
* 条件を満たすような区間を求める問題において利用可
*   「条件」について、条件を満たす区間が、いずれかの構造になっている必要がある。

1.   区間 [left, right) が「条件」を満たすなら、それに含まれる区間も「条件」を満たす
2.   区間 [left, right) が「条件」を満たすなら、それを含む区間も「条件」を満たす


<dequeを用いた実装>
```

for c in a:
    q.append(c)  ## dequeの右端に要素を一つ追加する。
    (追加した要素に応じて何らかの処理を行う)

    while not (満たすべき条件):
        rm = q.popleft() ## 条件を満たさないのでdequeの左端から要素を取り除く
        (取り除いた要素に応じて何らかの処理を行う)

    (何らかの処理を行う。
     whileがbreakしたので、dequeに入っている連続部分列は条件を満たしている。
     特に右端の要素から左に延ばせる最大の長さになっている。)
```

<通常の実装>

```

K = int(input())
r=0 #右端
ans=0

for l in range(len(S)):
  while r<=len(S)-1 and sum(l,r+1)<=K: #sum(l,r)はS[l]からS[r]までの和を返す関数
```

````python
#連続する部分列に含まれている要素がK種類以下で、最も長いものの長さ
from collections import deque,defaultdict
N,K = map(int,input().split())
A = list(map(int,input().split()))
d = defaultdict(int)
#尺取り法をdequeで！
que = deque()
ans = 0
for c in A:
  #print(que)
  que.append(c)
  d[c]+=1
  while que and (not len(d)<=K):
    rm = que.popleft()
    if d[rm]>=2: #消す要素が２個以上のときは、辞書に登録されている数を-1する
      d[rm]-=1
    elif d[rm]==1:#消すものについて、登録されている個数が1個のときは、辞書から消す
      d.pop(rm)
  ans = max(ans,len(que))
print(ans)
````

## 座標圧縮

````python
#関数
def compress(arr):
    *XS, = set(arr)
    XS.sort()
    return {e: i for i, e in enumerate(XS)}
````

## ランレングス圧縮(RLE)

* 連続したデータを一つ分のデータと連続した長さで表現する

例：AAABBCCCCA  → A3B2C4A1

* 注意:データによっては圧縮後のデータが元データよりも大きくなってしまう

<使用方法>
1. 与えられた文字列で連続する回数・個数に着目する問題のときに使う(連続した部分をまとめて扱う)

2. 変化する部分のみに着目する(連続部分の境界部のみに着目した考察など)
3. データサイズを圧縮する

````python
def rle(s): #ランレングス圧縮した文字列を返す
    tmp, count, ans = s[0], 1, ""
    for i in range(1,len(s)):
        if tmp == s[i]:
            count += 1
        else:
            ans += tmp+str(count)
            tmp = s[i]
            count = 1
    ans += tmp+str(count)
    return ans

#groupby関数をそのまま使う(同じキーを持つ連続するグループをまとめて(キー、グループ)の組にしてイテレータで返す)
from itertools import groupby
def rle2(listed):#リストをランレングス圧縮する
  return [[key,len(list(group))] for key,group in groupby(listed)]   
````

````python
def rle_num(s): #ランレングス圧縮したときの数値部分のリストを返す
  n = len(s)
  block = []
  num = 0
  for i in range(n):
      num += 1
      if i < n-1 and s[i] != s[i+1]:
          block.append(num)
          num = 0
  else:
      block.append(num)
  return block
````

## ダブリング

**全体の要素数がN個あって一回移動したときにどの要素に到達するのか定まっているとき、「*K個先の要素を求めるのにO(K)かかる*」ような状況**において、
* 前処理O(NlogK)
* クエリO(logK)
で行うことが出来るようにするアルゴリズム

<アルゴリズム>

ダブリングによるK個先の要素の求め方：

* 前処理：「doubling[k][i] : i 番目の要素から 2^k 先の要素は何か」を以下の式を利用して計算

 doubling[k+1][i] = doubling[k][doubling[k][i]]
* クエリ：前処理した結果を利用して K 個先の要素を求める

 現在地を now として、K を2進数として見た時の全ての桁について以下を行う

 K の k 桁目 が 1 ならば now = doubling[k][now] とする


````python
#ABC167D
#テレポーターをK回使うと、どの街に到着するか
N,K = map(int,input().split())
A = list(map(int,input().split()))

#dp[p][i]:i番目のマスから2^p回移動した先のマス
m = 100
dp = [[-1]*N for _ in range(m)]#10**18=2**60

#dp[0][i]には最初の移動先がそれぞれ入る
for i in range(N):
  dp[0][i] = A[i]-1

#ダブリングによりdp[k][i]：i番目の要素から2^k先の要素はなにかについて前処理する
for i in range(m-1):
  for j in range(N):
    #2^k先の要素がわかっていれば2^k先の要素の2^k先を求めることが出来る
    dp[i+1][j] = dp[i][dp[i][j]]

#現在地をnowとする
now = 0 #0-indexに注意
for i in range(m):
  #Kを２進数として見たときの全ての桁について、Kのk行目が1ならば
  if K>>i & 1: #一桁目から右ビットシフトを+1回ずつ行って各桁が1がどうか見る
    #now = doubling[k][now]である
    now = dp[i][now]

print(now+1) #i-indexに戻し忘れ注意
````

#### 二部グラフ判定

グラフが連結している場合、１つの頂点の色さえ決まれば、グラフを青と赤で塗る方法は１つに定まる

→工夫することで**深さ優先探索で解ける**

1.   すべての頂点を白色で塗る
2.   一番最初に頂点１を訪問し、頂点１を青色で塗る

３の操作を繰り返す

3.   隣接する白色頂点がない：一歩戻る　　　　　　　　　　　　　　　　　　　　　　　　　隣接する白色頂点がある：これらの中で番号が最小の頂点を訪問する。新たに頂点を訪問する際には、前の頂点と異なる色で塗る
4.   最終的にどの隣り合う２頂点も異なる色で塗られていたら、グラフは二部グラフである。

````python
from collections import deque
#from collections import defaultdict

#頂点N、辺の数がM
N,M = list(map(int,input().split()))

#グラフの入力
G = [[] for i in range(N+1)]

for _ in range(M):
    A,B = map(int,input().split())
    G[A].append(B)
    G[B].append(A)

#頂点を探索したかどうかの管理(頂点の個数分配列を用意)
color = [0 for i in range(N+1)]

bipartite = True

#すべての頂点について見ないと、グラフによっては探索しきれない可能性がある(ひたすらのWAの原因)
for s in range(1,N+1):
  if color[s]==0:
    color[s]=1
    que = deque([s])
    #キューの要素がなくなるまで処理を繰り返す
    while len(que)>0:
        p = que.popleft()#直近で追加したグラフの頂点を取得
        for q in G[p]:#結合しているグラフの頂点を参照
            if color[q] == 0:#塗られていないなら別の色で塗る(違う色は-colorで交互に塗れる)
                color[q] = -color[p]
                que.append(q) #キューに移動先の頂点を入れる
            elif color[q] == color[p]:#同じ色だったら2部グラフではないと返し終了させる
                bipartite = False
                break

print('Yes' if bipartite else 'No')
````

````python
import sys
sys.setrecursionlimit(10**9)

m = int(input())
n = int(input())
grid = [list(map(int,input().split())) for _ in range(n)]

ans = 0
#dfs中に今の探索が何回目かの情報がほしい→引数として変数を持つ！！！（重要）
def dfs(x,y,times):
    global ans
    ans = max(ans,times)
    #二度と同じ場所を探索できないように置き換える（置き換えるタイミングに注意！）
    grid[y][x] = 0
    dirc = [(0,1),(0,-1),(1,0),(-1,0)]
    for dx,dy in dirc:
        nx,ny = x+dx,y+dy
        if not (0<=nx<m and 0<=ny<n):
            continue
        if grid[ny][nx]==1:
            dfs(nx,ny,times+1)
            #print(times)
    #帰りがけ処理はここから行う！
    #探索抜けた後は、grid[y][x]=1として解除しておく→異なるパスでは再度訪問しても問題がない、
    #というより、そうしないと最大値を求められない
    grid[y][x] = 1

for h in range(n):
    for w in range(m):
        if grid[h][w]==1:
            dfs(w,h,1)

print(ans)
````

### 幅優先探索

**幅優先探索**

*   出発地点に近い頂点から順番に調べる
*   キューを用いて実装

**解ける問題**
*   単一始点最短経路問題（重みありの場合は、ダイクストラ法で）


**実装の方法**
*   始点に隣接している頂点から順に探索していく
*  **頂点を探索したかどうかの管理、探索予定の頂点を格納する配列**の２点が重要

#### 単一始点最短経路問題：重みなし

````python
# キューをインポート
from collections import deque

# 入力を受け取る。座標の調整のため、スタート地点とゴール地点の座標を-1する。
R, C = map(int, input().split())
sy, sx = map(int, input().split())
gy, gx = map(int, input().split())

sy -= 1
sx -= 1
gy -= 1
gx -= 1

# 迷路の情報を配列Gで受け取る
G = [input() for _ in range(R)]

# キューをQに入れ、スタート地点を追加(探索予定の頂点をキューに格納する)
Q = deque()
Q.append([sy, sx]) #スタート始点を最初に探索

# 未訪問と始点からの距離を管理するdistを定義。スタート地点に0を代入。
dist = [[-1]*C for _ in range(R)]
dist[sy][sx] = 0

# 今回は移動する４方向を事前に用意した。
dirc = [(0, 1), (0, -1), (1, 0), (-1, 0)]

# キューの要素がなくなるまで処理を繰り返す。
while len(Q) > 0:
  y, x = Q.popleft() #popleft()関数を使用して、キューから最初の要素を取り出すとともに削除する

# 移動先で繰り返し処理
  for dy, dx in dirc:
    y2 = y + dy
    x2 = x + dx

    # 移動先が迷路の範囲内でなければ、continue(ループ1周して次のループに入る)
    if not (0 <= y2 < R and 0 <= x2 < C):
      continue

    # 移動先が壁だったら、continue
    if G[y2][x2] == "#":
      continue

    # 移動先が未訪問なら移動前の距離＋１をdistに入れて、キューに移動先の座標を追加
    if dist[y2][x2] == -1:
      dist[y2][x2] = dist[y][x] + 1
      Q.append([y2, x2])

# ゴールの距離を出力
print(dist[gy][gx])
````

#### 最短経路に行く道の本数

v から遷移できる各 v'に対して次を行う
* v 'に未到達なら dist[v'] を dist[v]+1 に更新し、cnt[v'] に cnt[v] を代入する
*v ′に到達済みかつ dist[v']=dist[v]+1 なら、cnt[v'] に cnt[v] を加算する
* そうでないなら何もしない

(通常なら)
最短距離を求めるBFSでは、最短距離を表す配列をdist として,
* v'に未到達ならdist[v']をdist[v]+1に更新
* そうでないなら何もしない


<わかりやすいイメージ>
スタートからゴールまでの最短距離が3だとします。このとき、答えは 『**ゴールに繋がっている**最短距離が2の頂点への最短経路の組み合わせ数』の合計

````python
from collections import deque
N,M = map(int,input().split())
G = [[]*N for _ in range(N)]
for _ in range(M):
  a,b = map(int,input().split())
  a -= 1
  b -= 1
  G[a].append(b)
  G[b].append(a)

# キューをQに入れ、スタート地点を追加(探索予定の頂点をキューに格納する)
Q = deque()
Q.append(0) #スタート始点を最初に探索

# 未訪問と始点からの距離を管理するdistを定義。スタート地点に0を代入。
dist = [-1]*N
dist[0] = 0

#最短経路の数を保存
cnt = [0]*N
cnt[0] = 1

#幅優先探索
while len(Q):
  city = Q.popleft()
  for c in G[city]:
    if dist[c]==-1:
      dist[c] = dist[city]+1
      cnt[c] = cnt[city]
      Q.append(c)
    else:
      if dist[c]==dist[city]+1:
        cnt[c] += cnt[city]

print(cnt[N-1]%(10**9+7))
````

### ダイクストラ法

* 単一始点最短経路問題：重み付き をO(N^2)で解くことが出来る(ただし、辺の重みが非負のときに限る)
* 優先度つきキューを使うことでO((V+E)log(V))で求められる

````python
def dijkstra(G,start):
  from heapq import heappush,heappop 
  """
  ダイクストラ法(優先度付きキュー使用)
  from heapq import heappop,heappushの使用を忘れずに！
  引数：グラフ（重み付き）、スタート地点
  返り値：頂点の個数分の配列を返し、スタートから配列のindexまでの最短距離を求めることが出来る
  """
  N = len(G)
  done = [False]*N
  #距離の管理
  INF = float('inf')
  dist = [INF]*N #スタート地点以外の値は∞で初期化
  
  #スタート地点の重み（距離）
  dist[start] = 0 #スタートは0で初期化
  pq = [(0,start)] #ヒープには(その頂点へのコスト：頂点)の情報が入っている
  
  while pq:
    #ヒープから取り出し
    cost,v = heappop(pq)
    if done[v]:
      continue
    done[v] = True
    #最もコストが小さい頂点を探す
    for v,d in G[v]:#G[u]にはつながっている頂点番号とそこへのコストが入っている
      new_cost = cost + d
      #更新条件
      if dist[v] > new_cost: 
        dist[v] = new_cost
        heappush(pq,(new_cost,v)) #pqに(new_cost,v)
  return dist
````

*   自分自身は０

### クラスカル法

最小全域木問題・・・複数都市があって、ある都市とある都市を結ぶ道路をつくるためには○○円必要といった形式の情報がいくつか与えられるとき、最小の金額でどの都市間も行き来可能にする方法を求める方法

全域木・・・Gに含まれるすべての頂点とGに含まれる辺の一部からなる木のこと

最小全域木・・・Gの全域木の中で辺の重みの和が最小である木のこと

**クラスカル法について**
*   辺のコストが小さい順に使用できる辺を使う貪欲法のアルゴリズム
*   全頂点がバラバラで辺のない状態からスタートする
*   辺のコストで辺をソートして、コストが小さい順に辺を一つずつ見ていき、この辺が結ぶ2つの頂点がまだつながっていなければ、この辺を使用して2つの頂点を連結させる
*   既に同じ連結成分であれば、この辺を使う意味はないので使わない



#### 最小全域木問題

````python
#UnionFindクラスを利用する
V, E = map(int,input().split())    # Vは頂点数、Eは辺数
es = []

for _ in range(E):
    s, t, c = map(int,input().split())
    es.append([c, s, t])    # c(辺のコスト)でソートできるように最初の要素にしておく

def kruskal():
    es.sort()    # 辺のコストが小さい順にソートする
    UFT = UnionFindTree(V)    # Union-Findの初期化
    res = 0
    for i in range(E):
        e = es[i]
        if not UFT.same(e[1], e[2]):
            UFT.unite(e[1], e[2])
            res += e[0]
    return res

res = kruskal()

print(res)
````

##### UnionFindクラス

````python
# Union-Find木のコード

class UnionFindTree():
    # n要素で初期化
    def __init__(self, n):
        self.n = n
        self.par = list(range(n))
        self.rank = [0] * n

    # 木の根を求める
    def find(self, x):
        if self.par[x] == x:
            return x
        else:
            self.par[x] = self.find(self.par[x])
            return self.par[x]

    # xとyの属する集合を併合
    def unite(self, x, y):
        x = self.find(x)
        y = self.find(y)
        if x == y:
            return

        if self.rank[x] < self.rank[y]:
            self.par[x] = y
        else:
            self.par[y] = x
            if self.rank[x] == self.rank[y]:
                self.rank[x] += 1

    # xとyが同じ集合に属するか否か
    def same(self, x, y):
        return self.find(x) == self.find(y)
````

### 木DP

根を一つ固定した木について、以下のように部分木に関する値を利用して計算するDP

**dp[v]:頂点vを根とする部分木について何らかの値**


木構造、特に根付き木の構造を持つ問題において、葉から根または根から葉の方向に順番に処理が進んでいく場合は、深さ優先探索による実装がうまくいく場合が多い。

### トポロジカルソート(DAGに対して)

**トポロジカルソート**
* 閉路のないグラフにおいて、辺に沿って進むほど値が大きくなるようにN個の頂点に0,...,N-1の番号を割り当てること
* **与えられた有向グラフに対し、各頂点を辺の向きに沿うように順序付けて並び替えること**
* **深さ優先探索における再帰関数を抜けた順番に頂点を並べ、それを逆順に並べ直す**ことでトポロジカルソートが得られる
* 計算量は(|V|+|E|)
* 何が嬉しいか→DPの更新の順序がわかる(DPはある状態を処理するときはその状態に遷移する状態は全て処理しきっている必要がある）
* DAGの最短経路はDPで！
* トポロジカルソートの番号にしたがって順番に値を計算することで、この問題をDPで解くことが出来る
* 有向パスのうち最大のもの→ある頂点cuについての遷移は全ての子供toについて
max(dp[to],dp[cu]+1)

**<アルゴリズム>**

1. 空のリストSを用意する
2. 入次数（有向グラフにおいて、頂点に入ってくる辺数）が0の頂点が存在するとき
* 入次数が0の頂点を1つ選びuとする←番号が最小のものを選ぶことで辞書順のリストを得られる
* uを1. で作った空のリストに追加する
* uおよびuから出ている辺をグラフから削除する
3. アルゴリズムが終了した時にグラフに頂点が残っていなければトポロジカルソートした結果となる

````python
#DAGの有向パスのうち最長のもの
#length[i]:iを始点とするバスのうち、最も長いものの長さ
def topologicalsort(N, G):
    import heapq
    in_number = [0] * N #各頂点の枝の本数をカウントするためのリストを用意する
    
    for i in G:
        for v in i:
            in_number[v] += 1 #頂点vの入次数をカウントする

    S = [] #手順1:空のリストを用意する。
    queue = [i for i in range(N) if in_number[i] == 0] #入次数が0の頂点を記録するためのリストを用意
    heapq.heapify(queue) # リストを優先度付きキューへ
    
    while queue:
        u = heapq.heappop(queue) #手順2.1：入次数が0の頂点でかつ最小の頂点uを取り出す
        S.append(u) #手順2.2：手順1で作ったリストSに追加しています。
        for u2 in G[u]:
            in_number[u2] -= 1 #手順2.3：uから出ている辺（枝）を削除
            if in_number[u2] == 0: #uから出ている辺を削除していく作業で頂点u2の入次数が0となった場合
                heapq.heappush(queue, u2) #u2はuの候補となる
    
    return S
#入力    
N,M = map(int,input().split())

G = [[] for _ in range(N)]
for _ in range(M):
  a,b = map(int,input().split())
  G[a-1].append(b-1)

#トポロジカルソートの順番を出力する
ans = topologicalsort(N,G)

#トポロジカルソート順にDP
#dp[i]:有向パスの最後の頂点がiのときのパスの最長の長さ
dp = [0]*N
#トポロジカルソート順に頂点を調べる
for v in ans:
  #つながっている頂点について
  for c in G[v]:
    #つながっている部分の長さの最大値(dp[i])と現在地点＋１(dp[v]+1)のmaxでdp[i]更新
    dp[c] = max(dp[c],dp[v]+1)

print(max(dp))
````

### 強連結成分分解

* 有向グラフにおいて、「お互いに行き来できる⇔同じグループ」を満たすように頂点をグループ分けすること
* 簡単に言うと、全ての強連結成分を一つの頂点に潰すことであり、DAGが出来る
* どんな有向グラフでも強連結な頂点のグループに分けることが出来る
* 計算量O（N+M）
<何が出来るの？>
* 非DAGをDAGにしてDPなり何なりできる！

<アルゴリズム>
1. 適当な頂点から深さ優先探索を行う，その際各頂点 v に対して頂点 v から進めなくなった順番 t(v)を格納する

2. グラフの矢印を全て逆向きにしたものに対して深さ優先探索を行う。その際 t(v)が大きい頂点からスタートする。行き止まったところまでを一つの連結成分とする。

[アルゴリズムについてより詳しくはこちら](https://manabitimes.jp/math/1250)

````python
# 強連結成分分解(SCC): グラフGに対するSCCを行う
# 入力: <N>: 頂点サイズ, <G>: 順方向の有向グラフ, <RG>: 逆方向の有向グラフ
# 出力: (<ラベル数>, <各頂点のラベル番号>)
def scc(N, G, RG):
    order = []
    used = [0]*N
    group = [None]*N
    def dfs(s):
        used[s] = 1
        for t in G[s]:
            if not used[t]:
                dfs(t)
        order.append(s)
    def rdfs(s, col):
        group[s] = col
        used[s] = 1
        for t in RG[s]:
            if not used[t]:
                rdfs(t, col)
    for i in range(N):
        if not used[i]:
            dfs(i)
    used = [0]*N
    label = 0
    for s in reversed(order):
        if not used[s]:
            rdfs(s, label)
            label += 1
    return label, group

# 縮約後のグラフを構築
def construct(N, G, label, group):
    G0 = [set() for i in range(label)]
    GP = [[] for i in range(label)]
    for v in range(N):
        lbs = group[v]
        for w in G[v]:
            lbt = group[w]
            if lbs == lbt:
                continue
            G0[lbs].add(lbt)
        GP[lbs].append(v)
    return G0, GP
````

### ネットワークフロー




＜基礎用語＞

辺連結度：スタートとゴールを決めたとき、互いに辺を共有しないパスの最大本数

辺素：互いに辺を共有しないこと

カット：頂点集合Vを(S,T)と分割すること

カットセット：S側に始点を持ちT側に終点をもつような辺の集合のことをカット(S,T)に関数カットセットという

カットの容量：カットセットに含まれる辺の本数

<最小カット問題>

* 辺連結度に関する問題の強双対性

 辺素なs-tパス集合の最大本数＝s-tカットのうち容量が最小のもの(すべての辺が容量１と考えられる)

<最大流問題>

* 辺に容量成約がついた有向グラフを考えるとき、始点から終点へ、各辺の容量成約を守り最大いくらまで終点へ流せることができるか

* 最大流最小カット定理

 **最大フローの総流量＝s-tカットの最小容量**

* フォード・ファルカーソン法 
 
 1. ネットワークフローから残余ネットワークを作成する

 2. 残余ネットワークにおいて、SからTへ向かうルートがあるか探索する。ルートが存在しない場合は、終了する。

 3. 残余ネットワークを用いて、SからTへの最大流量を求める

 4. 3の結果を反映させた新しいネットワークフローを作成し、1の処理に戻る







````python
from collections import deque
 
 
class Dinic:
    def __init__(self, n):
        self.n = n
        self.links = [[] for _ in range(n)]
        self.depth = None
        self.progress = None
 
    def add_link(self, _from, to, cap):
        self.links[_from].append([cap, to, len(self.links[to])])
        self.links[to].append([0, _from, len(self.links[_from]) - 1])
 
    def bfs(self, s):
        depth = [-1] * self.n
        depth[s] = 0
        q = deque([s])
        while q:
            v = q.popleft()
            for cap, to, rev in self.links[v]:
                if cap > 0 and depth[to] < 0:
                    depth[to] = depth[v] + 1
                    q.append(to)
        self.depth = depth
 
    def dfs(self, v, t, flow):
        if v == t:
            return flow
        links_v = self.links[v]
        for i in range(self.progress[v], len(links_v)):
            self.progress[v] = i
            cap, to, rev = link = links_v[i]
            if cap == 0 or self.depth[v] >= self.depth[to]:
                continue
            d = self.dfs(to, t, min(flow, cap))
            if d == 0:
                continue
            link[0] -= d
            self.links[to][rev][0] += d
            return d
        return 0
 
    def max_flow(self, s, t):
        flow = 0
        while True:
            self.bfs(s)
            if self.depth[t] < 0:
                return flow
            self.progress = [0] * self.n
            current_flow = self.dfs(s, t, float('inf'))
            while current_flow > 0:
                flow += current_flow
                current_flow = self.dfs(s, t, float('inf'))
 
# 使い方
mf = Dinic(6)
mf.add_link(0, 1, 10)
mf.add_link(0, 3, 4)
mf.add_link(1, 2, 9)
mf.add_link(1, 4, 6)
mf.add_link(2, 5, 8)
mf.add_link(3, 4, 3)
mf.add_link(4, 5, 4)
print(mf.max_flow(0, 5))
````

### べき乗高速

````python
def pow_k(x, n):
    """
    O(log n)
    """
    if n == 0:
        return 1

    K = 1
    while n > 1:
        if n % 2 != 0:
            K *= x
        x *= x
        n //= 2

    return K * x
````

### 素数判定

````python
#１ 素数を判定する関数
def isprime(N):
    if N < 2:
        return False
    i = 2
    while i * i <= N:
        if N % i == 0:
            return False
        i += 1
    return True

isprime(101)
````

````python
#2
# 1 以上 N 以下の整数が素数かどうかを返す
def Eratosthenes(N):
    # テーブル
    isprime = [True] * (N+1)

    # 0, 1 は予めふるい落としておく
    isprime[0], isprime[1] = False, False

    # ふるい
    for p in range(2, N+1):
        # すでに合成数であるものはスキップする
        if not isprime[p]:
            continue

        # p 以外の p の倍数から素数ラベルを剥奪
        q = p * 2
        while q <= N:
            isprime[q] = False
            q += p

    # 1 以上 N 以下の整数が素数かどうか
    return isprime

# 50 以下の素数をすべて求める
N = int(input())
isprime = Eratosthenes(N)
index = [i for i,x in enumerate(isprime) if x==True]
print(*index)
````

### 約数

````python
# N の約数をすべて求める関数
def calc_divisors(N):
    # 答えを表す集合
    res = []

    # 各整数 i が N の約数かどうかを調べる
    for i in range(1, N + 1):
        # √N で打ち切り
        if i * i > N:
            break
        
        # i が N の約数でない場合はスキップ
        if N % i != 0:
            continue

        # i は約数である
        res.append(i)

        # N ÷ i も約数である (重複に注意)
        if N // i != i:
            res.append(N // i)

    # 約数を小さい順に並び替えて出力
    res.sort()
    return res
    
N = int(input())

A = calc_divisors(N)
for a in A:
  print(a)
````

### 素因数列挙

````python
#2 以上 N 以下の整数のうち、K 種類以上の素因数を持つものの個数

#整数を小さい方から見ていって、素数だったらその倍数の素因数の種類数に+1する
#その整数を見たときに素因数の種類数が0だったら素数
N,K = map(int,input().split())

cnt_prime = [0]*(N+1)
ans = 0
for x in range(2,N+1):
  # xが素数だったら
  if cnt_prime[x] == 0:
    #xの倍数yすべてに+1
    for y in range(x,N+1,x):
      cnt_prime[y] += 1
  #xの素因数の種類数がk以上だったら答えに+1    
  if cnt_prime[x]>=K:
    ans += 1
print(ans)
````

### 素因数分解

````python
"""nを素因数分解"""
"""2以上の整数n => [[素因数, 指数], ...]の2次元リスト"""

def factorization(n):
    arr = []
    temp = n
    for i in range(2, int(-(-n**0.5//1))+1):
        if temp%i==0:
            cnt=0
            while temp%i==0:
                cnt+=1
                temp //= i
            arr.append([i, cnt])

    if temp!=1:
        arr.append([temp, 1])

    if arr==[]:
        arr.append([n, 1])

    return arr

factorization(2019) 

## [[2, 3], [3, 1]] 
##  24 = 2^3 * 3^1
````

````python
#素因数分解
def prime_factorize(n):
    a = []
    while n % 2 == 0:
        a.append(2)
        n //= 2
    f = 3
    while f * f <= n:
        if n % f == 0:
            a.append(f)
            n //= f
        else:
            f += 2
    if n != 1:
        a.append(n)
    return a

N = int(input())
A = prime_factorize(N)
print(*A)
````

### 平衡三進法展開
＜豆知識＞

* 任意の整数は2ベキの和で表せる
* 引き算を考慮すると、任意の整数は3ベキの和差を用いると効率よく表せる

<解法>

繰り返し処理するだけ
たとえば 25 の場合

* 25 ÷ 3 = 8 ... 1 なので、答えに 1 を push して、25 から 1 引いて 3 で割る ({1}, 8)
* 8 ÷ 3 = 2 ... 2 なので、答えに -1 を push して、25 に 1 足して 3 で割る ({-1, 1}, 3}
* 3 ÷ 3 = 1 なので、答えに 0 を push して、3 を 3 で割る ({0, -1, 1}, 1)
* 1 ÷ 3 = 0 ... 1 なので、答えに 1 を push して、1 から 1 引いて 3 で割る ({1, 0, -1, 1}, 0)
* 0 になったので終了

````python
#解法１
from collections import deque
N = int(input())
ans = deque()
now = 1
while(N):
  if N%3==1:
    ans.append(now)
    N -= 1
  elif N%3==2:
    ans.append(-now)
    N += 1
  now *=3
  N = N//3
print(len(ans))
print(*ans)
````

````python
#解法2
def f(x):
    if x == 0:
        return []
    ret = []
    if x%3 == 0:
        ret = [i*3 for i in f(x//3)]
    if x%3 == 1:
        ret = [i*3 for i in f((x-1)//3)]
        ret.append(1)
    elif x%3 == 2:
        ret = [i*3 for i in f((x+1)//3)]
        ret.append(-1)
    return ret
ans = f(N)
print(len(ans))
print(' '.join(map(str,ans)))
````

### マンハッタン距離

マンハッタン距離→チェビシェフ距離の変換を行う

チェビシェフ距離とは

*   (**x1,y1),(x2,y2)間のチェビシェフ距離は max(|x1-x2|,|y1-y2|)**
*   つまり、x座標の差の絶対値か,y座標の差の絶対値かどちらか大きい方を距離とする定義
*   全点の座標を45度回転させるのと同義
*   変換するには**x'=x-y , y'=x+y**をする
*   マンハッタン距離とチェビシェフ距離は一致する
*   ある点において、他のすべての点とのチェビシェフ距離について最大のものは、変換後のx',y'の最大値を用いてmax(x-minx,y-miny,maxx-x,maxy-y)である（考えれば分かる）

````python
N,Q = map(int,input().split())
#全体を45度回転させた座標
cx = []
cy = []
for _ in range(N):
  x,y = map(int,input().split())
  #チェビシェフ距離への変換
  cx.append(x-y)
  cy.append(x+y)
#cx,cyの最大、最小それぞれ求める  
cx_max = max(cx)
cx_min = min(cx)
cy_max = max(cy)
cy_min = min(cy)
#Q個のクエリに答える
for _ in range(Q):
  q = int(input())
  q -= 1
  X,Y = cx[q],cy[q]
  ans = max(X-cx_min,Y-cy_min,cx_max-X,cy_max-Y)
  print(ans)
````
